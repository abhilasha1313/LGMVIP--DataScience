### Importing Required Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

### Loading Image

```python
image = cv2.imread("C://Users//Abhilasha//Pictures//brownie .jpeg")
```

### Converting to RGB format

```python
image2 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image2)
plt.axis(False)
plt.show()
```

### Converting to Grayscale

```python
grey_scale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_scale_image)
```

### Inverting Image

```python
inverted_image = cv2.bitwise_not(grey_scale_image)
# the bitwise NOT function flips pixel values. 
# All pixels that are greater than zero are set to zero
# and all pixels that are equal to zero are set to 255
plt.imshow(inverted_image)
```

### Blurring Image

```python
blurred_image = cv2.GaussianBlur(inverted_image, (111,111), 0)
plt.imshow(blurred_image)
```

### Inverting Blurred Image 

```python
inverted_blur_image = cv2.bitwise_not(blurred_image)
plt.imshow(inverted_blur_image)
```

### Pencil Sketch by Combining Grayscale and Inverted Blur

```python
sketch_image = cv2.divide(grey_scale_image, inverted_blur_image, scale=256.0)
plt.imshow(sketch_image)
```

### Saving Image 

```python
cv2.imwrite("sketch.png", sketch_image)
# to save the image
```

### BRG format of Pencil Sketch

```python
plt.figure(figsize = (10,10))
image3 = cv2.cvtColor(sketch_image, cv2.COLOR_RGB2BGR)
plt.imshow(image3)
plt.axis(False)
plt.show()
```
